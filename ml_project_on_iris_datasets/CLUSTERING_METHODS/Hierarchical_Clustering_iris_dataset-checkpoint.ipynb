```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
```

```python
X=iris.data
```

```python
X.shape
scale=StandardScaler()
labels1=scale.fit_transform(X)
print(labels1)
```

```python
pca=PCA(n_components=2)
labels2=pca.fit_transform(labels1)
print(labels2)
```

```python
plt.scatter(labels2[:,0],labels2[:,1])
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
plt.figure(figsize=(10,8))
Z = linkage(labels2, method='ward')
dendrogram(Z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python

agglo = AgglomerativeClustering(n_clusters=2, linkage='ward')
agglo.fit_predict(labels2)
agglo.labels_
```

```python
plt.scatter(labels2[:,0],labels2[:,1],c=agglo.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
score = silhouette_score(labels2, agglo.labels_)
print("Silhouette Score:", score)
```
