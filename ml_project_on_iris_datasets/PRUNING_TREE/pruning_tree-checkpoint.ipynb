```python
X=df.drop(["species"],axis=1,inplace=False)
y=iris.target
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
#pre pruning
parameter={
    "criterion":["gini","entropy","log_loss"],
    "splitter":["best","random"],
    "max_depth":[1,2,3,4,5],
    "max_features":["None","sqrt","log2"],
    # "ccp_alpha":[1,2,3,4,5,6,7]
}
from sklearn.model_selection import GridSearchCV
treemodel=DecisionTreeClassifier()
grid=GridSearchCV(treemodel,param_grid=parameter,cv=5,scoring="accuracy")
model=grid.fit(X_train,y_train)
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
from sklearn.metrics import accuracy_score,classification_report
accuracy=accuracy_score(y_pred,y_test)
print(accuracy)
```

```python
print(classification_report(y_pred,y_test))
```

```python
# plt.figure(figsize=(12,6))
# tree.plot_tree(model,filled=True)
plt.figure(figsize=(12,6))
tree.plot_tree(grid.best_estimator_,filled=True)
plt.show()
```

```python
print("Train:", grid.best_estimator_.score(X_train, y_train))
print("Test: ", grid.best_estimator_.score(X_test, y_test))
```
