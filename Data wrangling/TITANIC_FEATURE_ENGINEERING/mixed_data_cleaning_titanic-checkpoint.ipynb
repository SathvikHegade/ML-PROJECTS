```python
import pandas as pd
import numpy as np

df=pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df["numerical_data"]=df["number"].str.extract("(\d+)")
df["numerical_data"]=pd.to_numeric(df["numerical_data"], downcast="integer", errors='coerce')
df["numerical_data"]=df["number"].where(df["numerical_data"].notna())
df.head()
```

```python
df["categorical_data"]=df["number"].where(df["numerical_data"].isna())
df.head()
```

```python
df["numerical_cabin"]=df["Cabin"].str.extract("(\d+)")
df["numerical_cabin"]=pd.to_numeric(df["numerical_cabin"], downcast="integer", errors='coerce')

df["categorical_cabin"] = df["Cabin"].str[0]

df.head()
```

```python
df["numerical_data"]=pd.to_numeric(df["numerical_data"], downcast="integer", errors='coerce')
```

```python
df["numerical_data"]=df["numerical_data"].fillna(df["numerical_data"].mean())
df["numerical_data"]
df.head()
```

```python
df["categorical_data"]=df["categorical_data"].fillna("unavailable")
df.head()
```

```python
df["numerical_cabin"]=df["numerical_cabin"].fillna("unavailable")
```

```python
df["categorical_cabin"]=df["categorical_cabin"].fillna("unavailable")
df["Cabin"]=df["Cabin"].fillna("unavailable")
```

```python
df.head()
```

```python
df["numerical_data"] = df["numerical_data"].round(1)
```

```python
df.to_csv("mixed_data_cleaning_titanic.csv")
```
