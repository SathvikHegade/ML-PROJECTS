```python
from sklearn.datasets import fetch_california_housing
```

```python
import pandas as pd
cali=fetch_california_housing()
cali
```

```python
X=pd.DataFrame(cali.data,columns=cali.feature_names)
y=cali.target
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
error=r2_score(y_pred,y_test)
error
```

```python
param_grid = {
    'max_depth': [3, 8],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

```

```python
regress=DecisionTreeRegressor()
```

```python

from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(estimator=regress, param_grid=param_grid,cv=3, n_jobs=-1, scoring="r2")
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
error=r2_score(y_pred,y_test)
error
```

```python
from sklearn import tree
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
```

```python
tree.plot_tree(grid.best_estimator_,filled=True)
plt.show()
```

```python
print("Train:", grid.best_estimator_.score(X_train, y_train))
print("Test: ", grid.best_estimator_.score(X_test, y_test))
```
